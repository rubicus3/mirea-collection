# Задание 1.3.1

Условие

```
Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и
единицами
```



````python
import numpy as np

a = []
for i in range(8):
  for j in range(8):
    a.append((i + j + 1) % 2)

res = np.array(a).reshape(8,8 )
print(res)
````

# Задание 1.3.4


Условие

```
Создать матрицу с 0 внутри, и 1 на границах
```



````python
n = int(input("Введите размер матрицы: "))
a = np.full(n*n, 0).reshape(n, n)

for i in range(n):
  for j in range(n):
    if(i == 0 or i == n - 1 or j == 0 or j == n - 1):
      a[i][j] = 1

print(a)
````

# Задание 3.3.2

Условие


```
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.
```



````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'

dataframe = pd.read_csv(url)

s1 = MinMaxScaler()
s2 = StandardScaler()

dataframe[['sepal_length_cm']] = s1.fit_transform(dataframe[['sepal_length_cm']])
dataframe[['sepal_width_cm']] = s2.fit_transform(dataframe[['sepal_width_cm']])
dataframe
````
